```python
from datetime import date, timedelta
import matplotlib.pyplot as plt
import pandas as pd
import random
```

```python
sdate = date(2019,3,22)   # start date
edate = date(2019,4,9)   # end date
```

```python
dates = pd.date_range(sdate,edate-timedelta(days=1),freq='d')
```

```python
dates_str = [z.strftime("%Y-%m-%d") for z in dates]

dict_for_pd = {}
for dat in dates_str:
    dict_for_pd[dat] = [random.randint(1,30) for z in range(5)]
```

```python
df = pd.DataFrame.from_dict(dict_for_pd, orient='index',columns=['A', 'B', 'C', 'D', 'E'])
df["E"] = [random.randint(60,90) for z in range(len(dates_str))]
df["B"] = [random.randint(60,90) for z in range(len(dates_str))]
```

```python
df.head()
```

```python
# create figure and axis objects with subplots()
fig,ax = plt.subplots()
# make a plot
ax.plot(df.index, df.A, color="red", marker="o")
ax.plot(df.index, df.B, color="red", marker="o")

# set x-axis label
ax.set_xlabel("year",fontsize=14)
# set y-axis label
ax.set_ylabel("A",color="red",fontsize=14)

# twin object for two different y-axis on the sample plot
ax2=ax.twinx()
# make a plot with different y-axis using second axis object
ax2.plot(df.index, df.E,color="blue",marker="o")
ax2.set_ylabel("E",color="blue",fontsize=14)
plt.show()
```
